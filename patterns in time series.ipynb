### Setup

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

```python
def plot_series(time, series, start=0, end=None, format='-'):
    plt.plot(time[start:end], series[start:end], format)
    plt.xlabel('Time')
    plt.ylabel('Series')
    plt.grid(True)
```

### Trends

    Create a time series which has a trend line which is moving upwards

```python
def trend(time, slope=0):
    return slope * time
```

```python
time = np.arange(5*365)
series = trend (time, 0.1)
plot_series(time, series)
plt.show()
```

### Seasonality

    Create a time series which has seasonality in it

```python
def seasonality (time, amplitude):
    #np.where (if (foo>bar), x, y)
    series = time % amplitude
    return series
```

```python
series = seasonality (time, 100)
plot_series(time, series)
plt.show()
```

### Trends & Seasonality

```python
def trends_seasonality (time, amplitude, trend):
    #np.where (if (foo>bar), x, y)
    series = time % amplitude + time * trend
    return series
```

```python
series = trends_seasonality (time, 100, 0.1)
plot_series(time, series)
plt.show()
```

### Random noise OR White Noise

```python
from random import randrange

def white_noise (time):
    series = np.random.randn(len(time))
    return series
```

```python
series = white_noise (time)
plot_series(time, series)
plt.show()
```
